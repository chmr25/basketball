```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn_str = f"postgresql://basketball:adrian@192.168.1.66:5432/dev"

df= pd.read_sql("select * from team_stats;", conn_str)
```

**Average points scored & retrieved by 2-point shots for each team**

```python

df['points scored (2p shots)'] = df['avg_2p_scored']
df['points retrieved (2p shots)'] = df['avg_2p_retrieved']
df = df.sort_values(by=['points scored (2p shots)'], ascending=True)
df.plot(kind='barh', y=['points retrieved (2p shots)', 'points scored (2p shots)'], x='name', title="Average points by 2-point shots",
        color=["#000000","#1668a5"], grid=True).legend(loc=(1.04,1))
```

**Average points scored & retrieved by 3-point shots for each team**

```python
df['points scored (3p shots)'] = df['avg_3p_scored']
df['points retrieved (3p shots)'] = df['avg_3p_retrieved']
df = df.sort_values(by=['points scored (3p shots)'], ascending=True)
df.plot(kind='barh', y=['points retrieved (3p shots)', 'points scored (3p shots)'], x='name', title="Average points by 3point shots",
        color=["#000000","#800000"], grid=True).legend(loc=(1.04,1))
```

**Average points scored & retrieved by Free Throws for each team**

```python

df['points scored (Free Throws)'] = df['avg_3p_scored']
df['points retrieved (Free throws)'] = df['avg_3p_retrieved']
df = df.sort_values(by=['points scored (Free Throws)'], ascending=True)
df.plot(kind='barh', y=['points retrieved (Free throws)','points scored (Free Throws)'], x='name', title="Average points by Free throws",
        color=["#000000","#455b55"], grid=True).legend(loc=(1.04,1))
```
